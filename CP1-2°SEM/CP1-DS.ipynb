```python
# Importando bibliotecas
import pandas as pd 
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Input dos arquivos csv e definindo os Df
netflix_table = pd.read_csv("netflix_titles.csv")
tomatoes_movies_table = pd.read_csv("rotten_tomatoes_movies.csv")
tomatoes_critic_table = pd.read_csv("rotten_tomatoes_critic_reviews.csv")
```

```python
# Vizualiza√ß√£o das tabelas
display(netflix_table.head())
display(tomatoes_movies_table.head())
display(tomatoes_critic_table.head())
```

```python
# Filtrando a tabela 'netflix_table'
netflix_table = netflix_table.loc[netflix_table['type'] == 'Movie']
display(netflix_table)
```

```python
# Juntando as tabelas 'netflix_table' e 'tomatoes_movies_table'
netflix_tomatoes_movie_table = pd.merge(netflix_table, tomatoes_movies_table, left_on='title', right_on='movie_title', how='inner')
display(netflix_tomatoes_movie_table)
```

```python
# Contar a frequ√™ncia de cada g√™nero
generos_frequencia = netflix_movie_critic_table['genres'].value_counts().head(20)

# Criar o gr√°fico de barras para mostrar a quantidade de filmes por g√™nero
plt.figure(figsize=(10,6))
plt.bar(generos_frequencia.index, generos_frequencia.values, color='skyblue', edgecolor='black')

# Adicionar t√≠tulo e r√≥tulos aos eixos
plt.title('Quantidade de Filmes por G√™nero', fontsize=14)
plt.xlabel('G√™nero', fontsize=12)
plt.ylabel('Quantidade de Filmes', fontsize=12)

# Girar os r√≥tulos do eixo X para melhor visualiza√ß√£o
plt.xticks(rotation=45)

# Ajustar layout para evitar sobreposi√ß√£o
plt.tight_layout()

# Exibir o gr√°fico
plt.show()
```

### 4 - Como a avalia√ß√£o m√©dia dos filmes de drama evoluiu ao longo dos anos?

```python
# Evolu√ß√£o da avalia√ß√£o m√©dia ao longo dos anos
avg_rating_by_year = drama_movies.groupby('release_year')['tomatometer_rating'].mean()

# Plotar o gr√°fico
plt.figure(figsize=(10, 6))
avg_rating_by_year.plot(kind='line')
plt.title('Evolu√ß√£o da avalia√ß√£o m√©dia de filmes de drama')
plt.xlabel('Ano')
plt.ylabel('Avalia√ß√£o M√©dia')
plt.show() 
```

O gr√°fico mostra a evolu√ß√£o da avalia√ß√£o m√©dia de filmes de drama ao longo dos anos. Para gerar este gr√°fico, foi realizado um agrupamento por ano, onde os filmes de drama foram agrupados por seu ano de lan√ßamento (release_year), e a m√©dia das avalia√ß√µes (tomatometer_rating) foi calculada para cada ano. Esse agrupamento foi feito com o c√≥digo drama_movies.groupby('release_year')['tomatometer_rating'].mean().
 
A plotagem do gr√°fico foi feita utilizando a fun√ß√£o plot(), que criou um gr√°fico de linha com as m√©dias das avalia√ß√µes ao longo do tempo. O gr√°fico foi personalizado com t√≠tulo e r√≥tulos para os eixos X e Y, sendo exibido com plt.show().
 
D√©cadas de 1950 a 1970: Observa-se um crescimento acentuado nas avalia√ß√µes m√©dias at√© o final da d√©cada de 1950, com alguns picos e quedas durante as d√©cadas de 1960 e 1970. O valor m√°ximo ocorre por volta de 1980.
 
D√©cada de 1980: H√° uma grande queda nas avalia√ß√µes no in√≠cio da d√©cada de 1980, seguida de uma r√°pida recupera√ß√£o e novos picos ao longo da d√©cada.
 
D√©cadas de 1990 a 2010: As avalia√ß√µes mostram uma tend√™ncia de maior volatilidade, com flutua√ß√µes frequentes. N√£o h√° um crescimento constante, mas per√≠odos de alta e queda intercalados.
 
P√≥s-2010: A partir de 2010, a avalia√ß√£o m√©dia parece estabilizar em torno de valores menores do que em d√©cadas anteriores, com pequenas varia√ß√µes, indicando uma poss√≠vel queda na recep√ß√£o cr√≠tica de filmes de drama.
