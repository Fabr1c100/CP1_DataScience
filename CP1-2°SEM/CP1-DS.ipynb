```python
# Importando bibliotecas
import pandas as pd 
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Input dos arquivos csv e definindo os Df
netflix_table = pd.read_csv("netflix_titles.csv")
tomatoes_movies_table = pd.read_csv("rotten_tomatoes_movies.csv")
tomatoes_critic_table = pd.read_csv("rotten_tomatoes_critic_reviews.csv")
```

```python
# Vizualização das tabelas
display(netflix_table.head())
display(tomatoes_movies_table.head())
display(tomatoes_critic_table.head())
```

```python
# Filtrando a tabela 'netflix_table'
netflix_table = netflix_table.loc[netflix_table['type'] == 'Movie']
display(netflix_table)
```

```python
# Juntando as tabelas 'netflix_table' e 'tomatoes_movies_table'
netflix_tomatoes_movie_table = pd.merge(netflix_table, tomatoes_movies_table, left_on='title', right_on='movie_title', how='inner')
display(netflix_tomatoes_movie_table)
```

```python
# Contar a frequência de cada gênero
generos_frequencia = netflix_movie_critic_table['genres'].value_counts().head(20)

# Criar o gráfico de barras para mostrar a quantidade de filmes por gênero
plt.figure(figsize=(10,6))
plt.bar(generos_frequencia.index, generos_frequencia.values, color='skyblue', edgecolor='black')

# Adicionar título e rótulos aos eixos
plt.title('Quantidade de Filmes por Gênero', fontsize=14)
plt.xlabel('Gênero', fontsize=12)
plt.ylabel('Quantidade de Filmes', fontsize=12)

# Girar os rótulos do eixo X para melhor visualização
plt.xticks(rotation=45)

# Ajustar layout para evitar sobreposição
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

### 4 - Como a avaliação média dos filmes de drama evoluiu ao longo dos anos?

```python
# Evolução da avaliação média ao longo dos anos
avg_rating_by_year = drama_movies.groupby('release_year')['tomatometer_rating'].mean()

# Plotar o gráfico
plt.figure(figsize=(10, 6))
avg_rating_by_year.plot(kind='line')
plt.title('Evolução da avaliação média de filmes de drama')
plt.xlabel('Ano')
plt.ylabel('Avaliação Média')
plt.show() 
```

O gráfico mostra a evolução da avaliação média de filmes de drama ao longo dos anos. Para gerar este gráfico, foi realizado um agrupamento por ano, onde os filmes de drama foram agrupados por seu ano de lançamento (release_year), e a média das avaliações (tomatometer_rating) foi calculada para cada ano. Esse agrupamento foi feito com o código drama_movies.groupby('release_year')['tomatometer_rating'].mean().
 
A plotagem do gráfico foi feita utilizando a função plot(), que criou um gráfico de linha com as médias das avaliações ao longo do tempo. O gráfico foi personalizado com título e rótulos para os eixos X e Y, sendo exibido com plt.show().
 
Décadas de 1950 a 1970: Observa-se um crescimento acentuado nas avaliações médias até o final da década de 1950, com alguns picos e quedas durante as décadas de 1960 e 1970. O valor máximo ocorre por volta de 1980.
 
Década de 1980: Há uma grande queda nas avaliações no início da década de 1980, seguida de uma rápida recuperação e novos picos ao longo da década.
 
Décadas de 1990 a 2010: As avaliações mostram uma tendência de maior volatilidade, com flutuações frequentes. Não há um crescimento constante, mas períodos de alta e queda intercalados.
 
Pós-2010: A partir de 2010, a avaliação média parece estabilizar em torno de valores menores do que em décadas anteriores, com pequenas variações, indicando uma possível queda na recepção crítica de filmes de drama.
 
É possível observar que houve um aumento significativo nas avaliações médias entre as décadas de 1950 e 1980, mas após isso, as avaliações começaram a variar mais, e em tempos mais recentes, uma ligeira tendência de declínio é observada.

O gráfico recebe um título explicativo com plt.title(), e os eixos são rotulados: o eixo X mostra a duração em minutos e o eixo Y as notas de avaliação. Finalmente, o gráfico é exibido usando plt.show(), facilitando a visualização da relação entre duração e avaliação.
 
É possível observar que:

Não parece haver uma correlação linear direta entre a duração dos filmes e a avaliação. Filmes de curta, média e longa duração recebem uma ampla gama de notas, desde baixas até muito altas.
Há menos filmes de duração muito longa, mas eles também apresentam uma variedade de avaliações, algumas com pontuação alta (acima de 80) e outras bem baixas. No entanto, muitos desses filmes tendem a ter uma pontuação mais alta.
Com isso concluo que não há uma relação evidente entre a duração de um filme de drama e sua avaliação. Tanto filmes curtos quanto longos podem obter boas ou más avaliações, sugerindo que outros fatores além da duração desempenham um papel mais significativo na qualidade percebida pelos críticos e pelo público.

Ele agrupa os filmes pelo ano de lançamento com `groupby('release_year').size()` para contar quantos filmes foram lançados por ano e plota esses dados em um gráfico de barras com `plot(kind='bar')`, ajustando o tamanho para 10x6 polegadas. O gráfico recebe um título explicando que ele mostra a tendência de lançamentos, com o eixo X representando os anos e o eixo Y a quantidade de filmes. Por fim, o gráfico é exibido com `plt.show()`, permitindo visualizar essa evolução ao longo do tempo.
 
A tencência é de que entre 2014 e 2018, houve um aumento constante na quantidade de lançamentos de filmes de drama, com o ano de 2018 atingindo o pico, com mais de 200 filmes lançados. A partir de 2019, há uma diminuição significativa no número de filmes lançados. Esse declínio é especialmente acentuado em 2020 e 2021, onde a quantidade de lançamentos é drasticamente menor. Isso pode ter ocorrido por conta do declínio abrupto em 2020 e 2021 pode estar relacionado à pandemia de COVID-19, que impactou fortemente a indústria cinematográfica global, com adiamentos e reduções na produção de filmes.

Em geral, apesar de algumas diferenças nas notas ao longo dos anos, é possível notar que, em muitos momentos, as opiniões de críticos e público se aproximam, especialmente nos períodos de maior estabilidade na produção cinematográfica.
